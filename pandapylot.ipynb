```python
import statsmodels.formula.api as smf
model = smf.ols(
    "exam_score ~ mental_health_rating + sleep_hours + social_media_hours + study_hours_per_day + exercise_frequency", 
    data=finn
).fit()

print(model.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
predictors = finn[["mental_health_rating", "sleep_hours", "social_media_hours", "study_hours_per_day", "exercise_frequency"]]
vif = pd.DataFrame({"Variable": predictors.columns, "VIF": [variance_inflation_factor(predictors.values, i) for i in range(predictors.shape[1])]})
print(vif)
```
